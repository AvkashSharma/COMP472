# Data Set 1 - Distribution 

```python
import numpy as np
import pandas as pd
import seaborn as sns

DS_train = pd.read_csv("dataset/train_1.csv", header=None)
DS_val = pd.read_csv("dataset/val_1.csv", header=None)
DS_test = pd.read_csv("dataset/test_with_label_1.csv", header=None)

df_train= pd.DataFrame()
df_train["Class"] = pd.Series(DS_train.iloc[:, -1].values)
sns.catplot(y="Class", kind="count",palette="ch:.25", data=df_train).set(title="Training Data Set 1")

df_val= pd.DataFrame()
df_val["Class"] = pd.Series(DS_val.iloc[:, -1].values)
sns.catplot(y="Class", kind="count",palette="ch:.25", data=df_val).set(title="Validate Data Set 1")

df_test= pd.DataFrame()
df_test["Class"] = pd.Series(DS_test.iloc[:, -1].values)
sns.catplot(y="Class", kind="count",palette="ch:.25", data=df_test).set(title="Testing Data Set 1")
```
